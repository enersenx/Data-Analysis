```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load(f):
    dataset = pd.read_csv(f,delimiter=",",names=['c1','c2','c3','c4','c5'])
    return dataset
```

```python
dataset = load("Dataset-Exos.txt")
print(dataset)
```

```python
def scatterplot(dataset, attr1, attr2):
    x = dataset[attr1]
    y = dataset[attr2]
    plt.scatter(x,y)
    plt.title('Scatter plot')
    plt.xlabel('values of x')
    plt.ylabel('values of y')
    plt.show()
```

```python
scatterplot(dataset, 'c1', 'c2')
```

```python
def histogram(dataset, attr):
    x = dataset[attr]
    plt.hist(x, bins=range(int(min(x)), int(max(x))+1), edgecolor='black')
    plt.title('Histogram')
    plt.xlabel('values of x')
    plt.show()
```

```python
histogram(dataset, 'c1')
```

```python
def box(dataset, attr):
    x = dataset[attr]
    plt.boxplot(x)
    plt.title('BoxPlot with outliers')
    plt.ylabel('values of x')
```

```python
box(dataset, 'c1')
```

```python
def box_no_outliers(dataset, attr):
    x = dataset[attr]
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    val = 1.5 * (q3 -q1)
    below_q1 = q1 - val
    print(below_q1)
    above_q3 = q3 + val
    print(above_q3)
    condition = (x >= below_q1) & (x <= above_q3) 
    x_without_outliers = x[condition] #keep only the values inside that meet the condition (between q1 and q3)
    print(x_without_outliers)
    
    plt.boxplot(x_without_outliers)
    plt.title('BoxPlot without outliers')
    plt.ylabel('values of x')
```

```python
dataset_outliers = load("Dataset-Exos - outliers.txt")
print(dataset_outliers)
```

```python
# shows the outliers
box(dataset_outliers, 'c1')
```

```python
# doesn't show the outliers
box_no_outliers(dataset_outliers, 'c1')
```
