```python

def centers(dataset,attr):
    #sort dataset by given column
    dataset_sorted = dataset.sort_values(by=attr)
    print(dataset_sorted)
    
    #calcule moyenne
    avg = 0
    cpt = len(dataset_sorted)
    column = dataset_sorted[attr]
    print("column: \n")
    print(column)
    
    for line in column:
        avg += line
    avg = avg/cpt
    
    #calcule mediane
    if(cpt%2 == 0):#pair
        print('even ',cpt)
        idx1 = cpt // 2
        print('idx1 ', idx1)
        idx2 = (cpt // 2) - 1
        print('idx2 ', idx2)
        med = (column.iloc[idx1] + column.iloc[idx2]) / 2
    else:
        print('odd')
        idx = cpt // 2
        med = column.iloc[idx]
    return avg, med
    
```

```python
avg, med = centers(dataset,"c1")
print('avg: ', avg, '...', 'med: ', med)
```

```python
def quartiles(dataset, attr):
    dataset_sorted = dataset.sort_values(by=attr)
    
    n = len(dataset_sorted)
    print('n ', n)
    
    column = dataset_sorted[attr]
    
    q0 = column.iloc[0]
    q4 = column.iloc[-1]
    
    # Calculate q2
    x,q2 = centers(dataset,attr)
    
    # calculate q1
    if n % 4 == 0:  # if divisible by 4 --> 1/4
        print('div by 4')
        q1_idx1 = n // 4
        print('q1_idx1 ', q1_idx1)
        q1_idx2 = (n // 4) - 1
        print('q1_idx2 ', q1_idx2)
        q1 = (column.iloc[q1_idx1] + column.iloc[q1_idx2]) / 2
    else:
        print('not div by 4')
        q1_idx = n // 4
        print('q1_idx ', q1_idx)
        q1 = column.iloc[q1_idx]
    
    # calculate q3
    if (3 * n) % 4 == 0:  # if 3 * n is divisible by 4 --> 3/4
        print('div3 by 4')
        q3_idx1 = (3 * n) // 4
        q3_idx2 = ((3 * n) // 4) - 1
        q3 = (column.iloc[q3_idx1] + column.iloc[q3_idx2]) / 2
    else:
        print('not div3 by 4')
        q3_idx = (3 * n) // 4
        print('q3_idx ', q3_idx)
        q3 = column.iloc[q3_idx]
    
    return q0, q1, q2, q3, q4

```

```python
q0, q1, q2, q3, q4 = quartiles(dataset, 'c1')
```

```python
print('q0: ', q0)
print('q1: ', q1)
print('q2: ', q2)
print('q3: ', q3)
print('q4: ', q4)
```
